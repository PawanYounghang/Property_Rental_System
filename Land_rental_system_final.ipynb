```python
# Land Rental System
# Define classes and data structures

class Land:
    def __init__(self, kitta_number, city, direction, anna, price, availability):
        self.kitta_number = kitta_number
        self.city = city
        self.direction = direction
        self.anna = anna
        self.price = price
        self.availability = availability

class Transaction:
    def __init__(self, land, customer, rent_date, duration):
        self.land = land
        self.customer = customer
        self.rent_date = rent_date
        self.duration = duration

# Define functions

# Read land data from the text file and create Land objects
def read_land_data(filename):
    # Read land data from the text file and create Land objects
    lands = []
    file = open(filename, 'r')
    next(file)
    for i in file:
        data = i.strip().split(',')
        kitta_number = int(data[0])
        city = data[1]
        direction = data[2]
        anna = int(data[3])
        price = int(data[4])
        availability = data[5]
        land = Land(kitta_number, city, direction, anna, price, availability)
        lands.append(land)
    return lands

def display_available_lands(lands):
    # Display the lands that are available for rent
    print("Available Lands:")
    for land in lands:
        if land.availability.lower() == "yes":
            print(f"Kitta Number: {land.kitta_number}, City/District: {land.city}, Facing: {land.direction}, Anna: {land.anna}, Price: {land.price}, Availability: {land.availability}")

def rent_land(lands):
    # Rent a land and generate a note/invoice
    print("Available lands:\n")
    index = 1
    available_lands = []
    for i in lands:
        if i.availability.lower() == 'yes':
            available_lands.append(i)
            print(f"{index}. {i.city} __ Available : {i.availability}, Anna: {i.anna}, Price - {i.price}")
            index+=1
        
    while True:
        choice = int(input("Choose an no. corresponding to the land:"))
        if 1<=choice<=len(lands):
            land = available_lands[choice - 1]
            land.availability = 'no'
            break
        else:
            print("Enter the valid number")
    customer = input("Name of the customer: ")
    rent_date = input("Date: ")
    duration = int(input("Duration for rent(in months): "))
    invoice = Transaction(land, customer, rent_date, duration)

    is_rental = False
    generate_note_invoice(invoice,is_rental)
    save_transaction(invoice, is_rental)
    

def return_land(lands):
    # Return a rented land and generate a note/invoice
    print("Rented lands:\n")
    index = 1
    non_available_lands = []
    for i in lands:
        if i.availability.lower() == 'no':
            non_available_lands.append(i)
            print(f"{index}. {i.city} __ Available : {i.availability}, Anna: {i.anna}, Facing - {i.direction}")
            index+=1
    
    while True:
        choice = int(input("Choose an no. corresponding to the land you want to return:"))
        if 1<=choice<=len(lands):
            land = non_available_lands[choice - 1]
            land.availability = 'yes'
            break
        else:
            print("Enter the valid number")
    customer = input("Name of the customer: ")
    rent_date = input("Date: ")
    duration = int(input("Duration for rent(in months): "))
    invoice = Transaction(land, customer, rent_date, duration)

    is_rental = True
    apply_fine(non_available_lands)
    generate_note_invoice(invoice,is_rental)
    save_transaction(invoice, is_rental)

def generate_note_invoice(transaction,is_rental):
    # Generate a note/invoice for a transaction
    obj = transaction.land
    if is_rental == False:
        print("\n\n ....................... \n\n")
        print("\t\t\033[1mTechnoPropertyNepal Pvt. Ltd.\033[0m\n\n")
        print(f"Customer name:\033[4m {transaction.customer}  \033[0m\t\t\t\tBill-no:____\n")
        print(f"Rent Date:\033[4m {transaction.rent_date} \033[0m")
        print(f"Rent Duration:\033[4m {transaction.duration} months \033[0m")
        print(f"City/District:\033[4m {obj.city} \033[0m\n\n")
        print(f"Sn.\tKitta No.\tDirection(facing)\tAnna\tPrice(per month)")
        print(f"1.\t{obj.kitta_number}\t\t{obj.direction}\t\t\t{obj.anna}\t{obj.price}\n\n")
        print(f"\033[4m                                     \033[0m\n\n")
        
    elif is_rental == True:
        print("\n\n ....................... \n\n")
        print("\t\t\033[1mTechnoPropertyNepal Pvt. Ltd.\033[0m\n\n")
        print(f"Customer name:\033[4m {transaction.customer}  \033[0m\t\t\t\tBill-no:____\n")
        print(f"Return Date:\033[4m {transaction.rent_date} \033[0m")
        print(f"Rent Duration:\033[4m {transaction.duration} months \033[0m")
        print(f"City/District:\033[4m {obj.city} \033[0m\n\n")
        print(f"Sn.\tKitta No.\tDirection(facing)\tAnna\tPrice(per month)")
        print(f"1.\t{obj.kitta_number}\t\t{obj.direction}\t\t\t{obj.anna}\t{obj.price}\n\n")
        if apply_fine == True:
            print(f"**Late fees of NRs. 10000 | will be added.")
        else:
            print("**no late fees for the rent..")
        print(f"\033[4m                                     \033[0m\n\n")

def save_transaction(transaction, is_rental):
    #file = open('return_transactions.txt','w', newline='\n')
    #file.write("Customer_name,return_date,Duration(in months),Kitta.no,City/District,Facing\n")
    # Save the transaction details to a file
    obj = transaction.land
    try:
        if is_rental == False:
            file = open('Booking_transactions.txt','a')
            file.write(f"{transaction.customer},{transaction.rent_date},{transaction.duration},{obj.kitta_number},{obj.city},{obj.direction}\n")
        elif is_rental == True:
            file = open('Return_transactions.txt','a')
            file.write(f"{transaction.customer},{transaction.rent_date},{transaction.duration},{obj.kitta_number},{obj.city},{obj.direction}\n")
    except:
        print("Error!")
    file.close()
    
def apply_fine(lands):
    import random
    apply_fine = random.choice([True,False])
    return apply_fine

def main():
    # Main program logic
    lands = read_land_data("land.txt")
    #display_available_lands(lands)

    while True:
        print("1. Rent a land")
        print("2. Return a land")
        print("3. Quit")

        choice = input("Enter your choice: ")

        try:
            if choice == "1":
                rent_land(lands)
                #display_available_lands(lands)
            elif choice == "2":
                return_land(lands)
            elif choice == "3":
                break
            else:
                print("Invalid choice. Please try again.")
                
        except:
            print("please enter the integer value")
    

# Run the main program
if __name__ == "__main__":
    main()
```
